Geometry managers:
------------------------------
- Tkinter provides 3 geometry methods that really helps with positioning our widgets

###### 1. pack():
- Used to organize widget in the block
- syntax:
    > widget.pack(options)
    - options:
        * expand
        * fill--tk.X,tk.Y,tk.BOTH
        * side

###### 2. grid():
- It is one of the geometry manager and organizes the widgets in the tabular form
- syntax:
   > widget.grid(options)
   * options:
       * column
       *columnspan
       *row
       *rowspan

###### 3. place():
- It is one of geometry manager and organize the widgets to specific x and y coordinates
- syntax:
    > widget.place(options)
- x horizontal offset pixels
- y vertical offset pixels
- height height of pixels
- width

### Widgets:
##### 1. Button
    - syntax:
        > w=Button(parent,options)
    - options:
        * activebackground
        * activeforeground
        * bd--border width in pixels
        * bg--background color of button
        * fg--foreground of button text
        * font--button text
        * height
        * image--set image display on the button
        * command--it is set to function call which is scheduled when the function is called
##### 2. Canvas
- Used to add structural graphics
- We can draw graphs and plot the python application
- Syntax:
    > w=Canvas(parent,options)

```python
# pack()

from tkinter import *
parent=Tk()

button1=Button(parent,text='submit',fg='skyblue',bg='brown')
button2=Button(parent,text='submit',fg='violet',bg='brown')
button3=Button(parent,text='submit',fg='blue',bg='brown')
button4=Button(parent,text='submit',fg='green',bg='brown')

button1.pack(side=TOP)
button2.pack(side=RIGHT)
button3.pack(side=BOTTOM)
button4.pack(side=LEFT)

parent.mainloop()
```

```python
# place()

from tkinter import *

top=Tk(className=' Registration')
top.geometry('400x300')
top.configure(bg='cyan')

name=Label(top,text='Name').place(x=50,y=40)
in1=Entry(top).place(x=100,y=40)

email=Label(top,text='Email').place(x=50,y=70)
in2=Entry(top).place(x=100,y=70)

pwd=Label(top,text='Password').place(x=50,y=100)
in3=Entry(top).place(x=120,y=100)

top.mainloop()
```

```python
# creating simple button with options

from tkinter import *
from tkinter import messagebox


top=Tk(className='Simple Button')
top.geometry('400x300')

def fun1():
    messagebox.showinfo('Pop-Up','blue button clicked')

def fun2():
    messagebox.showinfo('Pop-Up','brown button clicked')

b1=Button(top,text='Simple',activebackground='cyan',pady=10,font='Arial',activeforeground='blue',fg='cyan',bg='blue',command=fun1).pack(side=LEFT)
b2=Button(top,text='SimpleSecond',activebackground='brown',pady=10,font='Arial',activeforeground='green',fg='brown',bg='green',command=fun2).pack(side=LEFT)

top.mainloop()
```

```python
# Frame() - it is a container

import tkinter as tk
win=tk.Tk()

f1=tk.Frame(master=win,width=100,height=80,bg='red').pack(fill=tk.BOTH)
f2=tk.Frame(win,width=50,bg='orange').pack(fill=tk.X)
f3=tk.Frame(win,width=25,bg='green').pack(fill=tk.Y,expand=True)

win.mainloop()
```
