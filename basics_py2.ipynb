```python
'''

Regular Expression:
-------------------
-> If we want to represent a group of strings according to a particular pattern then we use "regular expressions"
    1. pattern of 10 digit mobile number
    2. pattern of email id
    
Main important application areas of RE's:
-----------------------------------------
1. form validation(email, mobile nmbr)
2. to develop pattern matching applications(unix -grep, windows- ctrl+f)
3. to develop digital circuits
4. to develop communation protocols TCP/IP,etc

-> We use 're' module to implement regular expression based application
-> it contains several in-built functions to use RE 

1. compile() function:
----------------------
-> It convert pattern into python object, this object is called 'RegEx' object

Note: Once RegEx object is ready , then we need to create matcher object

2. finditer() function:
-----------------------
-> used to find how many matchers are available

Note: Once we got matcher object, we can call multiple methods on that 
        1. start() - start index of the match
        2. end() - end+1 index of the match
        3. group() - returns matched string

Ex:
    pattern='ab' {'ab' is available or not in given string.
                   if available
                      - how many times available
                      - where it is available(index)
                 }
    target_string='abaabbabab'
-> we can add pattern inside 'finditer()' as argument
-> sometimes requirement may b either a or b or c
-> sometimes want to search for any alphabets or gitits or any symbols then we need to go for 'character classes'
Character classes:
------------------

1. [abc] --> it will take either a or b or c
2. [^abc] --> except 'abc' , remaining will be printed
3. [a-z] --> any lowercase alphabet symbols
4. [A-Z] --> any uppercase alphabet symbols
5. [0-9] --> any numerical values
6. [a-zA-Z] --> both lowercase and uppercase
7. [0-9a-zA-Z] --> for any alphanumeric
7. [^0-9a-zA-Z] --> for any special symbols

Predefined character classes:
-----------------------------
\s -- to search for space character (pattern,target string)
\S -- except spcae character
\d -- search for digit
\D -- except digits
\w -- search for any word character (alphanumeric)
\W -- any character except word (except alphanumeric)
.  -- every character (it may be alphanumeric/digits/special characters)

Quantifiers:
------------
-> It can be used to specify no.of occurances to match 

a  - exactly one 'a'
a+ - atleast one 'a'
a* - any no.of a's including zero a's also(zero or more)
a? - atmost one 'a' , either zero 'a' or one 'a'
a{n} - excatly 'n' no.of a's

Important functions in 're' module:-
-----------------------------------

1.match(): It is used to check the given pattern at beginning of the target string(should not check in middle/last)
    
    if available
        then return match object
    else
        otherwise "None"

2.fullmatch(): It will check the full string whether it is matched or not


3.search(): It will check whether the given pattern is present or not in the target string
    if matches
        match object of the first occurence
    else
        None

4.findall(): It will find all the occurences of  match & returns into list. It returns list object which contains all matches

5.finditer(): (see above)

6.sub(): It is used for substitution or replacement of pattern

7.subn(): To know the no.of replacements 
            Syntax:
                t=re.subn(regex,replacement,target string)

8.split(): For spliting the given string according to particular pattern. Return type of 'split()' is 'list'

9.compile(): (see above)


Regular Expression:
-------------------
All own language identifiers
1. The allowed characters are:
    - Alphabets
    - Digits
    - #
2. First character should be lower case alphabet symbol from a to k [a-k]
3. Second character should be any digit divisible by 3 [0369]
4. The length of identifier should be atleast 2


'''
```

```python
'''
Python program to extract all mobile numbers present in input.txt file and write output to output.txt file

'''
import re

f1=open('input.txt')
f2=open('output.txt','w')

for line in f1:
    list=re.findall('[6-9]\d{9}',line)
    for list1 in list:
        f2.write(list1+'\n')
print('extracted successfully...')
```

```python
'''
Multi Threaded Programming:
---------------------------
Multi Tasking: Capability of performaing multiple task simultaneously
                                            or
               It is ability of an OS to perform different tasks at same time

Thread: 
-------
- It is an independent flow of execution 
- A single process can consists of multiple threads
- Each thread in a program performs a particular task
- Each process has one thread i.e, always running. This is the "Main Thread"
- This main thread actually creates the child thread objects
- Child thread is also initiated by the main thread
- Thread is a python object

Types of Multi-tasking:-
-----------------------

1. Process Based Multi-tasking
------------------------------
- Executing several tasks simultaneously, where each task is a seperate independent process
    Ex: Typing a program in editor, Downloading something, Listening to music

2. Thread Based Multi-tasking
-----------------------------
- Executing several tasks simultaneously, where each task is an independent part of the same process
    
Note: Multi-threading in python can be achieved by importing 'threading' module
      Every program has one thread is always running is called "MultiThread"

Defined thread in python:
------------------------
we have 3 ways
1. creating thread without using any class
2. creating thread by extending thread class
3. creating thread without extending thread class


'''
```

```python
import threading

# this thread is always runs
print('Current executing thread:',threading.current_thread().getName())
```

```python
# creating thread without using any class

from threading import *
def display():
    for i in range(5):
        print('child thread',i) # by default child thread name is 'thread 1'
t=Thread(target=display)
t.start()

for i in range(5):
    print('main thread',i)
```

```python
# creating thread by extending thread class

from threading import *

class MyThread(Thread):
    def run(self): # run() (is a instance method) is a predefined method override thread class
        for i in range(5):
            print('child thread')

t = MyThread() # main thread creates thread object and start thread object
t.start()

for i in range(5):
    print('main thread')
```

```python
# creating thread without extending thread class

from threading import *

class test:
    def m1(self):
        for i in range(5):
            print('child thread-1')
obj=test()
t=Thread(target=obj.m1)
t.start()

for i in range(5):
    print('main thread')
```

```python
# to check the performance of program without using multithreading concept

import time

def doubles(n):
    for num in n:
        time.sleep(1)
        print('double value: ',2*num)


def square(n):
    for num in n:
        time.sleep(2)
        print('square value: ',num*num)
        
n=[1,2,3,4,5]

begin=time.time()
doubles(n)
square(n)
end=time.time()

print('begin time:',begin)
print('end time:',end)
print('actual time:',end-begin)
```

```python
# to check the performance of program using multithreading concept
# join(): if main thread wants to wait until completing some other thread, then we should go for join mathod

import time
from threading import *

def doubles(n):
    for num in n:
        time.sleep(1)
        print('\ndouble value: ',2*num)

        
def square(n):
    for num in n:
        time.sleep(1)
        print('\nsquare value: ',num*num)
        
n=[1,2,3,4,5]

begin=time.time()

t1=Thread(target=doubles,args=(n,))
t2=Thread(target=square,args=(n,))
t1.start()
t2.start()

t1.join()
t2.join()


end=time.time()

print('begin time:',begin)
print('end time:',end)
print('actual time:',end-begin)
```

```python
from threading import *
import time

def display():
    print(current_thread().name,'......started')
    time.sleep(1000)
    print(current_thread().name,'.....ended')
    
print('no.of active threads:',active_count())

t1=Thread(target=display,name='child thread-1')
t2=Thread(target=display,name='child thread-2')
t3=Thread(target=display,name='child thread-3')

t1.start()
t2.start()
t3.start()

print('current active threads:',active_count())
time.sleep(2000)
print('no.of active threads:',active_count())
```
